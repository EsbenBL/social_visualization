> * Let's make a jitter-plot (that is, code up something like **Figure 2-1** from DAOST from scratch), but based on *SF Police data*. My hunch from inspecting the file is that the police-folks might be a little bit lazy in noting down the **exact** time down to the second. So choose a crime-type and a suitable time interval (somewhere between a month and 6 months depending on the crime-type) and create a jitter plot of the arrest times during a single hour (like 13-14, for example). So let time run on the $x$-axis and create vertical jitter.

```python
### Exercise 3.1 ### 
from datetime import datetime
import pandas as pd
import seaborn as sns 
import numpy as np 
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Georgia"

# Load, specify columns and convert to datatime
report_data = pd.read_csv("../Datasets/Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv")
```

```python
# Subset based on crimetype and hour of the day 
crime_type = "VANDALISM"
single_crime_data = report_data.query(f"Category =='{crime_type}'").copy()
# From January until and including June (Half a year)
single_crime_data["Date"] = pd.to_datetime(single_crime_data.Date, format = "%m/%d/%Y")
single_crime_data = single_crime_data.query("Date >= '2010-01-01' and Date < '2010-07-01'")
# To datetime and get hours and minutes 
single_crime_data["Time"] = pd.to_datetime(single_crime_data["Time"], format="%H:%M")
single_crime_data["Hour"] = single_crime_data["Time"].dt.hour
single_crime_data["Minute"] = single_crime_data["Time"].dt.minute
# In the 11th hours (from 23-00) 
single_crime_data = single_crime_data.query("Hour == 23")
print("Number of report in the period: ",len(single_crime_data))
```

```python
jitter = np.random.uniform(-0.5, 0.5, size=len(single_crime_data))

plt.scatter(x = single_crime_data.Minute, y = jitter,
             marker = "o", color = "None", linewidth = 1, edgecolor = "black")

plt.title(label = f"Report distribution of {crime_type.title()} by the minute between 11pm and midnight", size = 12)
plt.xlabel("Minute")
plt.yticks([])
plt.show()
```

> * Last time, we did lots of bar-plots. Today, we'll play around with histograms (creating two crime-data based versions of the plot-type shown in DAOST **Figure 2-2**). I think the GPS data could be fun to see this way. 
>   * This time, pick two crime-types with different geographical patterns **and** a suitable time-interval for each (you want between 1000 and 10000 points in your histogram)
>   * Then take the latitude part of the GPS coordinates for each crime and bin the latitudes so that you have around 50 bins across the city of SF. You can use your favorite method for binning. I like `numpy.histogram`. This function gives you the counts and then you do your own plotting. 

```python
#### Exercise 3.2 #### 
crime_one = "BURGLARY"
crime_two = "VEHICLE THEFT"
crime_year = 2012

crime_data = report_data.query(f"Category == '{crime_one}' or Category == '{crime_two}'").copy()
crime_data["Year"] = pd.to_datetime(crime_data.Date, format = "%m/%d/%Y").dt.year
crime_data = crime_data.query(f"Year == {crime_year}")
crime_data = crime_data.rename(columns = {"Y":"latitude", "X":"longitude"})
```

```python
## Define bins using both crime categories 
## All but the last (righthand-most) bin are half-open, while the last bin is a closed interval
overall_freq_count, lat_bins = np.histogram(crime_data.latitude, bins = 50)
```

```python
## Use the bins from both crimes to get frequency count for each crime within those bins
burglar_crime_freq, _ = np.histogram(crime_data.query(f"Category == '{crime_one}'").latitude, bins = lat_bins)
car_theft_crime_freq, _ = np.histogram(crime_data.query(f"Category == '{crime_two}'").latitude, bins = lat_bins)
```

```python
# average bin-boundaries to align shape of frequency count and bin-array
aligned_lat_bins = [np.mean(lat_bins[ele:ele+2]) for ele in range(len(lat_bins)-1)]
```

```python
## Plot 
fig, ax = plt.subplots()

ax.bar(aligned_lat_bins, height = car_theft_crime_freq, width = 0.0019,
        color = "red", linewidth = 1, edgecolor = "black", alpha = 0.4, 
        label = "Car theft")

ax.bar(aligned_lat_bins, height = burglar_crime_freq, width = 0.0019,
        color = "skyblue", linewidth = 1, edgecolor = "black", alpha = 0.4,
        label = "Burglary")

_xticks = [round(lat, 2) for lat in aligned_lat_bins[::5]]
ax.set_xticks(_xticks)
ax.set_xticklabels(_xticks, rotation=0)

ax.set_xlabel("Latitude")
ax.set_ylabel("Reported Crimes")
plt.legend(loc = "upper left")
plt.title(f"Reported crimes in {crime_year} on {crime_one.lower()} and {crime_two.lower()} distributed on latitude")
plt.tight_layout()
plt.show()
```

- **Answer**: Because all but the last (righthand-most) bin are half-open, while the last bin is a closed interval, we need to extend the vector, $v$, with the maximum value + 1, in order for the maximum value to be represented in it's own bin (otherwise it would be binned together with the second-largest value). 
- The sum over bin counts is 5000.

> * Next up is using the plot-type shown in **Figure 2-4** from DAOST, but with the data you used to create Figure 2.1. There is not a single great way to create kernel density plots in Python. [Here](https://www.geeksforgeeks.org/density-plots-with-pandas-in-python/), you'll finde a pandas based strategy, but you can also use `gaussian_kde` from `scipy.stats` ([for an example, check out this stackoverflow post](https://stackoverflow.com/questions/4150171/how-to-create-a-density-plot-in-matplotlib)) or you can use [`seaborn.kdeplot`](https://seaborn.pydata.org/generated/seaborn.kdeplot.html). There is also another nice tutorial for KDE plots [here](https://towardsdatascience.com/histograms-and-density-plots-in-python-f6bda88f5ac0).

```python
fig, ax = plt.subplots()

kde_configs = {"bw_method":0.03}
kde_line_configs = {"alpha":0.5, "label":"KDE line"}

ax = sns.histplot(single_crime_data.Minute,
                  color = "None", 
                  edgecolor = "black", 
                  stat = "density",
                  kde=True, 
                  bins = 60, 
                  label = "Histogram",
                  kde_kws = kde_configs,
                  line_kws = kde_line_configs,
                  ax = ax)

ax.lines[0].set_color('crimson')

ax.set_xlabel("Minute of the hour")
ax.set_title("Report distribution by the minute between 11pm and midnight", size = 12)

plt.legend()
plt.tight_layout()
```

> * OK, we're almost done, but we need some box plots. Here, I'd like you to use the box plots to visualize fluctuations of how many crimes happen per day. We'll use data from the 15 focus crimes defined last week.

>   * Create a box-and whiskers plot showing the mean, median, quantiles, etc for all 15 crime-types side-by-side. There are many ways to do this. I like to use [matplotlibs's built in functionality](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html), but you can also achieve good results with [seaborn](https://seaborn.pydata.org/generated/seaborn.boxplot.html) or [pandas](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.boxplot.html).

```python
focuscrimes = set(['WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY', 'DISORDERLY CONDUCT'])
```

```python
# subset data to focus crimes 
focus_crime_data = report_data[report_data.Category.isin(focuscrimes)].copy()
focus_crime_data.Date = pd.to_datetime(focus_crime_data.Date, format = "%m/%d/%Y").copy()
# Count number of crimes pr. day within each category 
box_plot_data = focus_crime_data.groupby("Date").Category.value_counts().rename("Count").reset_index()[["Category", "Count"]]
```

```python
fig, ax = plt.subplots(figsize=(9,7))

# Plot boxplot
sns.boxplot(data = box_plot_data,
            y = "Count",
            x = "Category",
            showmeans=True,
            meanline=True,
            flierprops={"marker": "o", "markerfacecolor":"white", "label":"Outlier"},
            boxprops={"facecolor":"white", "alpha":0.6},
            medianprops={"color": "coral", "alpha":0.6, "label":"Median"},
            meanprops={"linestyle":"-", "color":"red", "label":"Mean"},
            ax = ax)

# Add legends 
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[:3], labels[:3], loc="best")

ax.axhline(y=0, linestyle="--", color = "gray", alpha = 0.5)

plt.xticks(rotation=90, ha='right')
plt.ylabel("Reported crimes pr. day", size = 12)
plt.xlabel("")
plt.tight_layout()
plt.show()
```

If we also include days where no replorts were filled as date entries of count=1:

```python
# group by Date and Category and Count entries
category_count_by_date = focus_crime_data.groupby(['Date', 'Category']).size().reset_index(name='Count')

# create a Date range to cover all possible dates
date_range = pd.date_range(start=focus_crime_data['Date'].min(), end=focus_crime_data['Date'].max(), freq='D')

# create a new DataFrame with all possible combinations of dates and categories
categories = focus_crime_data['Category'].unique()

dates_and_categories = [(_date, _cat) for _date in date_range for _cat in categories]
dates_and_categories_df = pd.DataFrame(dates_and_categories, columns=['Date', 'Category'])

# merge the new DataFrame with the grouped data to fill missing dates and categories
category_count_by_date = pd.merge(dates_and_categories_df, category_count_by_date, on=['Date', 'Category'], how='left')

# fill missing Count values with 0
category_count_by_date['Count'] = category_count_by_date['Count'].fillna(0)
```

```python
fig, ax = plt.subplots(figsize=(9,7))

# Plot boxplot
sns.boxplot(data = category_count_by_date,
            y = "Count",
            x = "Category",
            showmeans=True,
            meanline=True,
            flierprops={"marker": "o", "markerfacecolor":"white", "label":"Outlier"},
            boxprops={"facecolor":"white", "alpha":0.6},
            medianprops={"color": "coral", "alpha":0.6, "label":"Median"},
            meanprops={"linestyle":"-", "color":"red", "label":"Mean"},
            ax = ax)

# Add legends 
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[:3], labels[:3], loc="best")

ax.axhline(y=0, linestyle="--", color = "gray", alpha = 0.5)

plt.xticks(rotation=90, ha='right')
plt.ylabel("Reported crimes pr. day", size = 12)
plt.xlabel("")
plt.tight_layout()
plt.show()
```

>    * And the map should center on San Francisco's `lat` and `lon`.
